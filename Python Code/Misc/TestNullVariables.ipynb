```python
import numpy as np
# If unclean data is having predictions made on it, then this method will fail.

# find all cols with null values 
# look up their accident_index in the dataframe 
# then fill in the missing ones

# Going through every row and selecting it if there is a null element in it.
# Then we select acciddnrt indicies from these and select the unique ones/.
unique_accident_index = list(set(data[data.isnull().any(axis=1)]['accident_index']))

#for accident_index in unique_accident_index[0:5]:
    
def fill_null_indices(j):
    x = data.loc[data['accident_index'] == j]
    indices = x.index
    null_indices = x.index[x.isnull().any(axis=1)].tolist()
    not_null_index = 0
    for i in indices:
        if i not in null_indices:
            not_null_index = i
            break
    fill_data = data.iloc[not_null_index, -14:]
    for i in null_indices:
        data.iloc[i, -14:] = fill_data

#print(len(unique_accident_index))

for c, j in enumerate(unique_accident_index):
    fill_null_indices(j)
        
#data.loc[data['accident_index'] == unique_accident_index[j]]
```
